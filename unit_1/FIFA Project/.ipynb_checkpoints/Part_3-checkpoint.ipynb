```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./file_for_project/fifa21_training.csv")
```

#### STANDARIZE

```python
num_cols = data._get_numeric_data()
num_cols
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
for c in num_cols[:-1]:   
    data[c]=StandardScaler().fit_transform(data[c].values.reshape(-1, 1))
```

#### ENCODING

```python
cat_cols = data.select_dtypes(object).columns
cat_cols
```

```python
one_hot_data=pd.get_dummies(data[cat_cols], drop_first=True) 
one_hot_data.head()
```

```python
data=pd.concat([data, one_hot_data], axis=1)   
data.drop(columns=cat_cols, inplace=True)
data.head()
```

#### SPLIT

```python
X=data.drop(columns=['OVA'])
y=data['OVA']

print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.2, random_state=50)  #PORQUE 48,50 Y NO 29.000?
```
